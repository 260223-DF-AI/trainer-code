# Pandas Demonstration

Pandas provides a method that makes it easy to consume CSV content into a pandas DataFrame object, using pd.read_csv().

```python
import pandas as pd
data: pd.DataFrame = pd.read_csv("sample_data.csv")
print(data)
```

To start working with DataFrames, we can use the head() and tail() methods to get a window into the DataFrame's data. This is especially useful with DataFrames with too many rows of data to easily view at once.

```python
print("DataFrame Head")
print(data.head())
print("\n"*2)
print("DataFrame Tail")
print(data.tail())
```

In order to access an individual column, you may index the DataFrame similarly to dictionaries:

```python
city_column: pd.Series = data["Age"]
print(city_column)
```

Pandas provides a method to filter a DataFrame by using boolean expressions:

```python
data[data["Age"] > 24]
```

```python
# Pandas uses slightly different syntax for using column length
# .str is necessary to treat the column values as strings
data[data["Name"].str.len() < 5]
```

```python
# Along with filtering, we can also sort a DataFrame's rows
sorted_data: pd.DataFrame = data.sort_values(by=['Age', 'Name']) 
print(sorted_data)
```

Pandas is great for analyzing data, as it provides many analytical functions.

```python
print(f"Mean age: {data["Age"].mean()}")
print(f"Age sum: {data["Age"].sum()}")
```

You can also clean data in different ways with Pandas

```python
# Drop a row (Sorry Rich)
for x in sorted_data.index:
    if sorted_data.loc[x, "Age"] > 32:
        sorted_data.drop(x, inplace = True)

# Modify an entry
for x in sorted_data.index:
    if sorted_data.loc[x, "Name"] == "Ben":
        sorted_data.loc[x, "Name"] = "Benjamin"
```

Now we can easily write this sorted dataframe to a new CSV file

```python
sorted_data.to_csv("sorted_data.csv") 
```
